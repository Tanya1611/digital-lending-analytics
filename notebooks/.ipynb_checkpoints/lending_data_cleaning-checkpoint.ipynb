```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset
df = pd.read_csv("lendingclub_raw.csv", low_memory=False)
```

```python
# To clearly see all column names - storing in a list.
lst = df.columns.tolist()
lst
```

##### For business analysis, Borrower Profile and Loan Details are needed.

```python
columns_needed = ["loan_amnt",
                  "term",
                  "int_rate",
                  "grade",
                  "emp_length",
                  "home_ownership",
                  "annual_inc",
                  "loan_status",
                  "purpose",
                  "dti",
                  "fico_range_low",
                  "fico_range_high"]
```

```python
# Analysis to be done in a dataframe with selective columns & their data
df = df[columns_needed].copy()
```

```python
# After selecting relevant loan and borrower attributes, Exploratory checks are to be performed using info(), describe(),
# and Null analysis to understand data quality, distributions, and required cleaning steps.
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
# Dropping the duplicate rows
df.drop_duplicates()
```

##### 1st Attribute -> Loan Amount

```python
df["loan_amnt"].isnull().sum()
```

```python
# There are many null values in the loan amount & since it's a critical data we require for analysis, we are dropping the null rows
df=df.dropna(subset=["loan_amnt"])
```

```python
df.isnull().sum()
# Many null value rows got removed with it already.
```

##### 2nd Attribute -> Term

```python
df.term.unique()
```

```python
# Remove uneccessary space
df["term"] = df["term"].str.strip()
```

```python
# Remove months string from data
df["term"] = df["term"].str.replace(" months","").astype(int)
```

```python
# Rename the term attribute
df = df.rename(columns={"term":"term_in_months"})
```

```python
df["term_in_months"].unique()
```

##### 3rd Attribute -> Interest rate

```python
df["int_rate"].unique()
```

##### 4th Attribute -> Grade

```python
df.grade.unique()
```

##### 5th Attribute -> Employee Length

```python
df["emp_length"].isna().sum()
```

```python
df["emp_length"] = df["emp_length"].fillna("Unknown")
```

```python
df.info()
```

##### 6th Attribute -> Home Ownership

```python
df["home_ownership"].unique()
```

##### 7th Attribute -> Annual Income

```python
df["annual_inc"].unique()
```

```python
df["annual_inc"].isnull().sum()
```

```python
df["annual_inc"] = df["annual_inc"].fillna(df["annual_inc"].median())
```

##### 8th Attribute -> Loan Status

```python
df["loan_status"].unique()
```

```python
df["loan_status"].value_counts()
```

##### When applying for a loan, the system usually decides whether to accept it or reject it.

```python
df["decision"] = df["loan_status"].apply(lambda x: "Approved" if x in ["Fully Paid", "Current"] else "Rejected")
```

```python
df["decision"].value_counts()
```

```python
df.info()
```

##### 9th Attribute -> Purpose

```python
df["purpose"].unique()
```

##### 10th Attribute -> DTI

```python
df["dti"].isnull().sum()
```

```python
df["dti"]=df["dti"].fillna(df["dti"].median())
```

```python
df.dti.unique()
```

```python
df.info()
```

```python
df.isna().sum()
```

##### Creating Credit Score

```python
df["credit_score"] = (df["fico_range_low"] + df["fico_range_high"]) / 2
```

```python
df["credit_score"].describe()
```

##### Creating Risk Segment

```python
df["risk_segment"] = df.apply(lambda x: "Low Risk" if (x["credit_score"] >= 700) or (x["dti"] < 25) else "High Risk", axis = 1)
```

```python
df.info()
```

```python
df.head()
```

```python
df.info()
```

# Exploratory Data Analysis & Business Insights

### 1. Overall Loan Approval Rate

```python
df["decision"].value_counts(normalize=True)*100
```

```python
approval_counts = df["decision"].value_counts()

# Creating Bar Chart
plt.figure()
plt.pie(approval_counts, labels=approval_counts.index, autopct='%1.1f%%')
plt.title("Loan Approval Distribution")

plt.show()
```

#### The analysis of the loan approval rate shows that the platform approves 86.5% of applications, indicating its overall approach.

### 2. Factors Drive Approval

```python
# Credit Score Affect Approval
df.groupby(df["decision"])["credit_score"].mean()
```

##### Credit score alone does not significantly differentiate approved and rejected (or risky) borrowers, indicating that lending decisions and loan outcomes depend on multiple factors beyond credit score.

```python
# Let's check the distribution
df.groupby("decision")["credit_score"].describe()
```

##### While credit score is an important factor, it alone does not guarantee loan success. Some high-credit borrowers still fall into risky segments, indicating the need for multi-factor risk assessment.

```python
# Loan Approval Rate based on Credit Score, DTI, Annual Income
df.groupby("decision")[["credit_score","dti","annual_inc"]].mean()
```

```python
drivers = df.groupby("decision")[["credit_score","annual_inc","dti"]].mean()

# Creating subplots (1 row, 3 columns)
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Plot 1: Credit Score
drivers["credit_score"].plot(kind="bar", ax=axes[0])
axes[0].set_title("Credit Score by Decision")
axes[0].set_xlabel("")
axes[0].tick_params(axis='x', rotation=0)

# Plot 2: Annual Income
drivers["annual_inc"].plot(kind="bar", ax=axes[1])
axes[1].set_title("Income by Decision")
axes[1].set_xlabel("")
axes[1].tick_params(axis='x', rotation=0)

# Plot 3: DTI
drivers["dti"].plot(kind="bar", ax=axes[2])
axes[2].set_title("DTI by Decision")
axes[2].set_xlabel("")
axes[2].tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.show()
```

#### Approval decisions are influenced by a combination of credit score, income, and debt levels rather than a single factor.

### 3. Are Higher Loan Amount Riskier?

```python
df.groupby("decision")["loan_amnt"].mean()
```

```python
df.groupby("decision")["loan_amnt"].describe()
```

```python
# Creating Box Plot
sns.boxplot(x="decision", y="loan_amnt", data=df)
plt.title("Loan Amount Distribution By Decision")
plt.show()
```

##### The analysis shows that loan amount alone does not significantly determine borrower risk. Larger loans are not necessarily riskier, as lending decisions are influenced by borrower financial strength such as income and credit profile.

##### While loan amount alone does not indicate risk, the loan-to-income ratio provides a better measure of borrower burden and potential risk.

```python
df["loan_to_income"] = df["loan_amnt"]/df["annual_inc"]
```

```python
df["loan_to_income"].describe()
```

##### Need to look into the data that caused the infinite result

```python
df["loan_amnt"].describe()
```

```python
df.loan_amnt.sort_values(ascending=False).head()
```

```python
df.annual_inc.describe()
```

```python
df.annual_inc.sort_values(ascending=False).head()
```

```python
# Removing extreme outliers
df=df[df["annual_inc"]<50000]
```

##### Extreme income values distort financial ratios, so outliers were removed to ensure meaningful borrower comparisons.

```python
df.annual_inc.describe()
```

```python
# Since annual_inc is a denominator value in the calculation of ratio, it should not be 0 we will get inf as a result.
df["annual_inc"] = df["annual_inc"].apply(lambda x: df["annual_inc"].mean() if x==0 else x)
```

```python
df.annual_inc.describe()
```

```python
df["loan_to_income"] = df["loan_amnt"] / df["annual_inc"] *100
```

```python
df.loan_to_income.describe()
```

```python
df.groupby("decision")["loan_to_income"].mean()
```

```python
# Creating Box Plot
plt.figure()
sns.boxplot(x="decision", y="loan_to_income", data=df)
plt.title("Loan-to-Income Ratio by Decision")
plt.xlabel("")
plt.ylabel("Loan-to-Income (%)")
plt.show()
```

##### Loan amount alone is not a strong indicator of risk. However, when analyzed relative to borrower income, higher loan burdens can better explain risk patterns.

```python
df.groupby("decision")["loan_to_income"].mean().plot(kind="bar")
plt.title("Average Loan-to-Income Ratio by Decision")
plt.xticks(rotation=0)
plt.show()
```

### 4. Decision vs Risk

```python
pd.crosstab(df["decision"], df["risk_segment"], normalize="index")
```

```python
cross = pd.crosstab(df["decision"], df["risk_segment"], normalize="index")

# Creating Heatmap
plt.figure()
sns.heatmap(cross, annot=True, fmt=".2f")
plt.title("Decision vs Risk Heatmap")
plt.show()
```

#### A portion of approved borrowers fall into high-risk segments, while some low-risk borrowers are rejected, indicating potential inefficiencies in the lending decision process.

# Key Findings

• The lending platform shows a balanced approval strategy, with a significant proportion of applications being approved, indicating a moderate risk appetite.

• Loan approval decisions are influenced by a combination of financial factors such as credit score, income, and debt-to-income ratio, rather than any single variable.

• Loan amount alone does not significantly determine borrower risk; instead, borrower financial strength and repayment capacity play a more critical role.

• Analysis of decision versus risk segments reveals that some approved borrowers fall into high-risk categories, while some low-risk borrowers are rejected, indicating potential inefficiencies in lending decisions.
